```python
%pylab inline
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs

X, y = make_blobs( n_samples=1000, n_features=2, cluster_std = 4.5, centers=2, random_state=42  )

cm='Spectral'
plt.figure( figsize=(15,10) )
plt.scatter( X[:, 0], X[:, 1], c=y, cmap=cm, s=75, alpha=0.7 )
```

We can create whatever story we want for this data. Maybe it relates to the results from two biomarkers. Maybe it's economic data indicating a bear or bull market. In order to think like a data scientist, you need to know what question to ask, and make sure that your analysis reflects that question. For example, if your question is simply "are these samples different" we could calculate the mean, standard deviations, and run a two-sample t-test to see if these samples are different. Lets do that.

```python
from scipy.stats import ttest_ind

print "Red Cluster Mean: {}".format( np.mean( X[ y == 0 ], axis = 0 ) )
print "Blue Cluster Mean: {}".format( np.mean( X[ y == 1 ], axis = 0 ) )
print "Are these different? {}".format( ttest_ind( X[ y == 0 ], X[ y == 1 ] )[1] < .01 )
```

That was interesting, but hardly informative. We could see by eye that they were different. It was not a good question to pursue, and so we need a more complex question to pursue in order to get a Ph.D in The Science. We've really reduced the data to look like the following-- when you use aggregates, you throw away the most important information.

```python
import numpy as np
from matplotlib.patches import Ellipse

plt.figure( figsize=(20,10))
ax = plt.subplot(121)

mean, std = np.mean( X[y==0], axis=0 ), np.std( X[y==0], axis=0 )
ax.add_artist( Ellipse( mean, std[0], std[1], alpha=0.6, facecolor='r' ) )
ax.add_artist( Ellipse( mean, std[0]*2, std[1]*2, alpha=0.3, facecolor='r' ) )
ax.add_artist( Ellipse( mean, std[0]*3, std[1]*3, alpha=0.1, facecolor='r' ) )
mean, std = np.mean( X[y==1], axis=0 ), np.std( X[y==1], axis=1 )
ax.add_artist( Ellipse( mean, std[0], std[1], alpha=0.6, facecolor='b' ) )
ax.add_artist( Ellipse( mean, std[0]*2, std[1]*2, alpha=0.3, facecolor='b' ) )
ax.add_artist( Ellipse( mean, std[0]*3, std[1]*3, alpha=0.1, facecolor='b' ) )
plt.xlim(-20,20)
plt.ylim(-15,30)

plt.subplot(122)
plt.scatter( X[:, 0], X[:, 1], c=y, cmap=cm, s=75, alpha=0.7 )
```

Now, the general machine learning classification question is this: "If we have a set of data points, and a label corresponding to each point, can we use this information to predict the label of future points?" How would we pursue this question if someone gave you this dataset? 

Lets begin by transforming this question. How do we prove that we can predict the labels of future points? The easiest way seems to be to somehow show an error rate for predicting the labels of future points given the points that we already have. One can then use their judgement to determine if a certain error rate is low enough to say a classification is worthwhile, depending on the application. 

Lets start by splitting our data into parts, around 80% and around 20%. The 80% will be the data that we play with, and the 20% will be the data we pretend will come in in the future, which we're attempting to classify. A very simple technique would seem to be to classify every point the same way that its closest neighbor is classified. This is called K-Nearest-Neighbors (KNN), where K in this case is 1. 

```python
from sklearn.neighbors import KNeighborsClassifier as knn
from sklearn.cross_validation import *

m, n = X.shape

h = 0.02
xx, yy = np.meshgrid( np.arange( X[:,0].min()-1, X[:,0].max()+1, h ), 
                      np.arange( X[:,1].min()-1, X[:,1].max()+1, h ) ) 

X_explore, X_future, y_explore, y_future = train_test_split( X, y, test_size=0.2 )

clf = knn( n_neighbors=1, weights='distance' ) 
clf.fit( X_explore, y_explore )
Z = clf.predict( np.c_[xx.ravel(), yy.ravel()] ).reshape( xx.shape )

plt.figure( figsize=(15, 15) )
plt.contourf( xx, yy, Z, cmap=cm, alpha=0.4 ) 
plt.scatter( X_explore[:, 0], X_explore[:, 1], s=50, c=y_explore, cmap=cm, alpha=0.8 )
plt.text( -7, -10, "Accuracy = {:2.2}".format( clf.score( X_explore, y_explore ) ), fontsize=28 )
```

Wow, look at that accuracy! 100%! And look at those boundaries, they certainly are getting every point! Before we ship this project off claiming 100% accuracy, lets just, for the sake of thoroughness, try to test the future examples using this classifier, trained on this data!

```python
plt.figure( figsize=(12, 12) )
plt.contourf( xx, yy, Z, cmap=cm, alpha=0.4 ) 
plt.scatter( X_future[:, 0], X_future[:, 1], s=80, c=y_future, cmap=cm, alpha=0.8 )
plt.text( -7, -10, "Accuracy = {:2.2}".format( clf.score( X_future, y_future ) ), fontsize=28 )
```

Wow, that sucks. What happened? We used to have 100% accuracy, and now not so much! This will not do, it was likely David's fault. 

We fell into the problem of OVERFITTING. We fell into a trap that only people like David fall into, which is that we attempted to judge the quality of an algorithm by giving it some points, their labels, and then asked it to give the labels to us. Most algorithms should be able to do this! What we want to do is to give it new points that the algorithm has never encountered before, in order to ensure that the point which is being judged was not also used to train the algorithm!

The solution? Split our 'training' data into a training split and a test split. For our cases, lets train our algorithm using 75% of the data, and test the quality of the algorithm using 25% of the data. 

```python
def plot_knn( **kwargs ):
    plt.figure( figsize=(20,10))
    X_train, X_test, y_train, y_test = train_test_split( X, y, train_size=0.75 )
    clf = knn( **kwargs )
    clf.fit( X_train, y_train )
    Z = clf.predict( np.c_[xx.ravel(), yy.ravel()] ).reshape( xx.shape )
    plt.contourf( xx, yy, Z, cmap=plt.cm.RdBu_r, alpha=0.33 )
    
    plt.scatter( X_train[ :,0 ], X_train[ :,1 ], c=y_train, cmap=cm, alpha=0.33, s=50 )
    plt.scatter( X_test[ :,0 ], X_test[ :,1 ], c=y_test, cmap=cm, s=50 )
    
    plt.text( -12, -9, "Training Set Accuracy = {:2.2}".format( clf.score( X_train, y_train ) ), fontsize=18 )
    plt.text( -12, -11, "Test Set Accuracy = {:2.2}".format( clf.score( X_test, y_test ) ), fontsize=18 )

from IPython.html.widgets import interact, IntSliderWidget

a = IntSliderWidget( min=1, max=100, step=1, value=1 )
i = interact( plot, n_neighbors=a )
```

So this accuracy isn't great. It seems like we're returning to the 0.85 accuracy figure from before. Now lets run this on the future examples and see if we get a better fit.

```python
plt.figure( figsize=(12, 12) )
plt.contourf( xx, yy, Z, cmap='RdBu', alpha=0.5 ) 
plt.scatter( X_future[:, 0], X_future[:, 1], s=80, c=y_future, cmap='RdBu' )
plt.text( -12, -10, "Future Set Accuracy = {:2.2}".format( clf.score( X_future, y_future ) ), fontsize=28 )
```

This accuracy may not be the best, but the point is that the accuracy that you found in the train/test split was VERY CLOSE to the actual accuracy that you found. The accuracy on the future test examples just randomly happened to be higher, this does not always happen. Lets use a different algorithm to calculate the decision boundaries--a common machine learning algorithm called a Support Vector Machine. 

```python
from sklearn.svm import SVC

X_train, X_test, y_train, y_test = train_test_split( X, y, train_size=0.75 )
clf = SVC()
clf.fit( X_train, y_train )
plt.figure( figsize=(12, 12) )
Z = clf.predict( np.c_[xx.ravel(), yy.ravel()] ).reshape( xx.shape )
plt.contourf( xx, yy, Z, cmap=cm, alpha=0.5 ) 
plt.scatter( X_train[:, 0], X_train[:, 1], s=80, c=y_train, cmap=cm )
plt.text( -12, -10, "Train Set Accuracy = {:2.2}".format( clf.score( X_train, y_train ) ), fontsize=28 )
plt.text( -12, -12, "Test Set Accuracy = {:2.2}".format( clf.score( X_test, y_test ) ), fontsize=28 )
```

```python
import numpy as np

gammas = 10 ** np.arange( -8, 4, .5 )
plt.figure( figsize=(16, 8) )
test_scores = []
train_scores = []
for gamma in gammas:
    clf = SVC( gamma=gamma )
    clf.fit( X_train, y_train )
    train_scores.append( clf.score( X_train, y_train ) )
    test_scores.append( clf.score( X_test, y_test ) )
plt.semilogx( gammas, train_scores, 'b-', markersize=4, label="Training Scores" )
plt.semilogx( gammas, test_scores, 'r-', markersize=4, label="Testing Scores" )
plt.text( 2*10**-8, 0.7, "UNDERFITTING", fontsize=24 )
plt.text( 10**-3, 0.7, "GOOD FIT", fontsize=24 )
plt.text( 10**2, 0.7, "OVERFITTING", fontsize=24 )
plt.xticks( fontsize=20 )
plt.yticks( fontsize=20 )
plt.legend( loc=8, prop={'size':25}, markerscale=2 )
```

```python
clf = SVC( gamma=10**-4.5 )
clf.fit( X_train, y_train )
plt.figure( figsize=(12, 12) )
Z = clf.predict( np.c_[xx.ravel(), yy.ravel()] ).reshape( xx.shape )
plt.contourf( xx, yy, Z, cmap=cm, alpha=0.5 ) 
plt.scatter( X_train[:, 0], X_train[:, 1], s=50, c=y_train, cmap=cm, alpha=0.3 )
plt.scatter( X_test[:, 0], X_test[:, 1], s=50, c=y_test, cmap=cm, alpha=0.7 )
plt.scatter( X_future[:, 0], X_future[:, 1], s=50, c=y_future, cmap=cm )
plt.text( -12, -8, "Train Set Accuracy = {:2.2}".format( clf.score( X_train, y_train ) ), fontsize=28 )
plt.text( -12, -10, "Test Set Accuracy = {:2.2}".format( clf.score( X_test, y_test ) ), fontsize=28 )
plt.text( -12, -12, "Future Set Accuracy = {:2.2}".format( clf.score( X_future, y_future ) ), fontsize=28 )
```

The next focus I want to have is on a thing called "feature selection". Often times in machine learning, you're given not only a lot of data ,bbut very high dimensional data. Not all of these dimensions, or features, actually mean anything. Lets take a look at a sample dataset!

```python
from sklearn.datasets import make_classification

X, y = make_classification( n_samples=1000, n_features=2, n_informative=1, n_redundant=0, n_repeated=0, n_classes=2, n_clusters_per_class=1 )

plt.scatter( X[:,0], X[:,1], c=y, cmap=cm )
```

We can see that only one of these features seems like a good pick for classification! The other has a lot of overlap. We could simply remove these features by eye. We could just show the histograms of the data projected into one dimension to see that one is good and the other isn't. 

```python
plt.figure( figsize=(14,6) )
plt.subplot(121)
plt.hist( X[ y==0, 0 ], facecolor='r' )
plt.hist( X[ y==1, 0 ], facecolor='b' )
plt.subplot(122)
plt.hist( X[ y==0, 1], facecolor='r' )
plt.hist( X[ y==1, 1], facecolor='b' )
```

On higher dimensional datasets, you're going to want to use an algorithm though! These range from SelectKBest, SelectPercentile, UnivariateSelection, RecursiveFeatureElimination, or Forests of Trees. Lets use forests of trees.
