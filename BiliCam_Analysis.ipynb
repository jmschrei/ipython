# BiliCam Data Analysis

<a href="http://homes.cs.washington.edu/~mayank/BiliCam.pdf">BiliCam</a> is an app designed to monitor levels of jaundice in newborn babies, designed in the Patel lab at University of Washington wth Lilian de Greef as the lead author. Briefly, jaundice tests require meausirng levels of bilirubin in newborn babies, a task which can easily be done by expensive equipment in hospitals but not easily done once the newborn is discharged. This app is designed to be a monitoring service for that time, so that anyone can simply take a picture of their child and get an estimated level of Bilirubin and a suggestion as to if that level is dangerous given the age of the newborn.

This app requires several parts; the first is the computer vision expertise needed to segment images into their respective parts and collect relevant data from these segments, such as color channel mean and medians for the face of the newborn. Secondly, machine learning techniques are needed to analyze this data and both predict bilirubin levels and whether or not this level is dangerous.

## Initial Machine Learning

Lets first load up the typical suspects, watermark this run, and load up the relevant data.

```python
from matplotlib import pyplot as plt
from scipy.stats import pearsonr
import seaborn as sns
import pandas as pd
import numpy as np

%matplotlib inline
%load_ext watermark
%watermark -a "Jacob Schreiber" -d -m

data = pd.read_csv( 'BiliCam_dataset_clean.csv' )
```

Now we have a typical regression problem. We have a set of features, and we want to predict a bilirubin level from them. First we do feature selection using a LASSO regression minimizing SSE, then we use a forest of extremely randomized trees to do the actual regression using only the selected features.

```python
from sklearn.linear_model import Lasso, LinearRegression
from sklearn.cross_validation import LeaveOneOut
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import ExtraTreesRegressor

# Omit color-card features, use features of index > i
i = 71

# Use n data points, 102 total
n = 102

# Slice X according to the above parameters
X = data.values[:n, i:]

# Select y from the original data
y = data['TSB'].values[:n]

# Get feature names
names = data.columns[i:]

def LOOCV( X, y ):
    '''
    Perform LOOCV and return the prediction vector. This involves two steps:
        (1) Feature selection using a LASSO regression
        (2) Regression using a forest of extremely randomized trees
    '''

    clf = Lasso( fit_intercept=True, warm_start=True )
    unbiased_clf = ExtraTreesRegressor( n_estimators=300, max_depth=None )
    sum_fi = np.zeros(X.shape[1])
    y_hat = np.zeros(n)

    for train, test in LeaveOneOut( n=n ):
        clf.fit( X[train], y[train] )
        coef = np.abs( clf.coef_ ) > 0

        unbiased_clf.fit( X[train][:,coef], y[train] )
        y_hat[test] = unbiased_clf.predict( X[test][:,coef] )

        sum_fi += coef 
    
    return y_hat, sum_fi

# Run the LOOCV method once
y_hat, sum_fi = LOOCV( X, y )

# Print the features selected 
print "Selected Features:"
print "\t" + "\n\t".join( data.columns[i:][sum_fi > 0] )

# Store the results as a pandas dataframe for later visualization
results = pd.DataFrame({'Predicted': y_hat, 'True': y })
```

Now that we've done LOOCV to do the predictions, we can plot the predictions versus the true values of bilirubin in the patients, calculating the pearson correlation and corresponding p-value.

```python
with sns.axes_style("white"):
    sns.jointplot( "Predicted", "True", results, kind="kde", stat_func=pearsonr, color="#4CB391")
    print "BILICAM"
    print "SSE:", np.sum( ( y_hat - y ) ** 2 )
    print "MSE:", np.mean( np.abs( y_hat - y ) )
    print "RMSE:", np.sqrt( np.mean( ( y_hat - y ) ** 2 ) )
```

We should also compare this to TcB, which is the other alternative to TSB.

```python
with sns.axes_style("white"):
    sns.jointplot( "TcB", "TSB", data, kind='kde', stat_func=pearsonr, color="#713aa3" )
    print "TcB"
    print "SSE:", np.sum( ( data['TcB'] - y ) ** 2 )
    print "MSE:", np.mean( np.abs( data['TcB'] - y ) )
    print "RMSE:", np.sqrt( np.mean( ( data['TcB'] - y ) ** 2 ) )
```

## Model Analysis

Lets get into more of the nitty-gritty of what the results are telling us, using this regressor! The first thing to do is look at the feature importance determined by LASSO over all the folds. We can plot a bar chart of mean feature importance across all LOOCV runs.

```python
fi = pd.Series( sum_fi ) / n
fi.index = names
fi.sort( ascending=True )
with sns.axes_style("whitegrid"):
    fi.plot( kind='barh', figsize=(3, 15), color='c', alpha=0.5, width=1. )
```

Now lets look at the distribution of correlation and SSE you get using the identified classification method.

```python
n_runs = 50
corrs = np.zeros( n_runs )
sses = np.zeros( n_runs )

for i in xrange( n_runs ):
    y_hat, _ = LOOCV( X, y )
    sses[i] = ( ( y - y_hat ) **2 ).sum()
    corrs[i] = pearsonr( y, y_hat )[0]
```

```python
plt.figure( figsize=(10, 4) )
ax = plt.subplot(121)
plt.title( "SSE KDE" )
plt.xlabel( "SSE" )
plt.ylabel( "Density" )
sns.kdeplot( sses, ax=ax, shade=True )

ax = plt.subplot(122)
plt.title( "Correlation KDE" )
plt.xlabel( "Correlation" )
plt.ylabel( "Density" )
sns.kdeplot( corrs, ax=ax, shade=True )
```

By running this multiple times, we can see the effect of the randomness associated with building our classifier. If our classifier were entirely deterministic, it would produce the same result every run. We can use Maximum A Posteriori (MAP) estimates to get a more stable estimate for our SSE and correlation values.
